```python
import pandas as pd
url = "https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv"
dados = pd.read_csv(url)
```

```python
#1 Realizar a média da metragem para cara um dos bairros (Paulo)
dados.groupby("Bairro")["Metragem"].mean()
```

```python
# 2  2 formas de selecionar os dados por bairro (consultar os métodos na documentação do pandas) (Thiago)
df = dados["Bairro"].unique()
result_br=[]
result_m2=[]

for br in df:
    tem_imoveis =( dados["Bairro"] == br)
    imoveis_bairro = dados[tem_imoveis]
    mediam2 = imoveis_bairro["Metragem"].mean()
    
    result_br.append(str(br))
    result_m2.append(mediam2)
    
df2 = pd.DataFrame({"Bairro": result_br, "Media_M2": result_m2})
```

```python
df2.sort_values(by=["Bairro"])
```

```python
#3 - Explorar alguns gráficos na documentação e aplicar nas demais colunas do DF, tentar colocar alguma conclusão (Thiago)
df3 = pd.DataFrame(dados)
df3.groupby("Bairro").agg("mean")
df3.plot()
```

```python
#3

media = df3.groupby("Bairro")["Quartos","Banheiros","Vagas"].agg("mean")
media
```

```python
#3

media.sort_values(by=["Quartos","Banheiros","Vagas"], ascending=False) [:5].plot(kind='bar')
```

```python
#4 - Pegar outros estatísticas dos dados, média, mediana, mim, max (Vivi)

max = df3.groupby("Bairro")["Quartos","Banheiros","Vagas"].agg("max")

min = df3.groupby("Bairro")["Quartos","Banheiros","Vagas"].agg("min")

mediana = df3.groupby("Bairro")["Quartos","Banheiros","Vagas"].agg('median')
```
